```python
# Xa cc tr chi khng c tn, tn b li
df_temp = df_temp[df_temp['name'].notna()]
def fix_name_game(name):
    if name == name.upper() or name == name.lower() or 'http' in name:
        return np.nan
    new_name = re.sub(r'\((.*?)\)', '', name)
    if re.search(r'\W\s', name):
        new_name = name[:re.search(r'\W\s', name).span()[0]].strip()
        if len(new_name) > 50:
            return np.nan
    return new_name.strip()
df_temp['name'] = df_temp['name'].apply(fix_name_game)
df_temp = df_temp[df_temp['name'].notna()].reset_index(drop = True)
```

```python
# Xa cc ct rating, rating_top, ratings_count v to li ct mi da trn ct ratings
temp_list = df_temp['ratings'].apply(ast.literal_eval)

del_col_rate = ['rating', 'rating_top', 'ratings_count', 'ratings']
df_temp.drop(columns = del_col_rate, inplace = True)

def get_avg_top(list_rate):
    if len(list_rate) == 0:
        return {'top_rating': 0, 'avg_rating': 0,'count_rating': 0}
    top_rating = list_rate[0]['id']
    count = list_rate[0]['count']
    avg = top_rating * count
    sum = count
    if len(list_rate) != 1:
        for i in range(1, len(list_rate)):
            if list_rate[i]['count'] > count:
                top_rating = list_rate[i]['id']
                count = list_rate[i]['count']
            avg += list_rate[i]['id'] * list_rate[i]['count']
            sum += list_rate[i]['count']
    return {'top_rating': top_rating, 'avg_rating': round(avg/sum, 4), 'count_rating': sum}

se_rating = temp_list.apply(get_avg_top)
df_temp['top_rating'] = se_rating.apply(lambda x: x['top_rating'])
df_temp['avg_rating'] = se_rating.apply(lambda x: x['avg_rating'])
df_temp['count_rating'] = se_rating.apply(lambda x: x['count_rating'])
```

```python
# Thay gi tr nan trong ct added_by_status bng gi tr khc, sau  iu chnh nh dng
df_temp['added_by_status'].replace(np.nan, "{'yet': 0, 'owned': 0, 'beaten': 0, 'toplay': 0, 'dropped': 0, 'playing': 0}", inplace = True)
def format_status(d_status):
    format = {
        'yet': d_status['yet'] if 'yet' in d_status.keys() else 0,
        'owned': d_status['owned'] if 'owned' in d_status.keys() else 0,
        'beaten': d_status['beaten'] if 'beaten' in d_status.keys() else 0,
        'toplay': d_status['toplay'] if 'toplay' in d_status.keys() else 0,
        'dropped': d_status['dropped'] if 'dropped' in d_status.keys() else 0,
        'playing': d_status['playing'] if 'playing' in d_status.keys() else 0
    }
    return json.dumps(format)
df_temp['added_by_status'] = df_temp['added_by_status'].apply(ast.literal_eval).apply(format_status)
```

```python
folder_path = '/content/RAWG_Dataset'
creators_df = pd.read_csv(f'{folder_path}/creators.csv').rename(columns = {'id': 'creator_id'})
developers_df = pd.read_csv(f'{folder_path}/developers.csv').rename(columns = {'id': 'developer_id'})
publishers_df = pd.read_csv(f'{folder_path}/publishers.csv').rename(columns = {'id': 'publisher_id'})


def extract_game_ids(games_list):
    try:
        games = ast.literal_eval(games_list)
        if isinstance(games, list):
            return [game['id'] for game in games if isinstance(game, dict) and 'id' in game.keys()]
    except (ValueError, SyntaxError):
        return []
    return []


def expand_fact_table(fact_df, related_df, related_column, new_name, game_column='games'):
    related_df[game_column] = related_df[game_column].apply(extract_game_ids)
    related_df_exploded = related_df.explode(game_column).dropna(subset=[game_column])

    if related_column == 'creator_id':
        related_df_exploded['positions'] =  related_df_exploded['positions'].apply(ast.literal_eval)
        related_df_exploded = related_df_exploded[[related_column, 'name', game_column, 'positions']]
        related_df_exploded = related_df_exploded.groupby(game_column).apply(lambda x: x[[related_column, 'name', 'positions']].to_dict('records')).reset_index()
    else:
        related_df_exploded = related_df_exploded[[related_column, 'name', game_column]]
        related_df_exploded = related_df_exploded.groupby(game_column).apply(lambda x: x[[related_column, 'name']].to_dict('records')).reset_index()
    related_df_exploded.rename(columns = {0: new_name}, inplace = True)
    related_df_exploded[new_name] = related_df_exploded[new_name].apply(lambda x: json.dumps(x))

    expanded_df = fact_df.merge(related_df_exploded, left_on='game_id', right_on = game_column, how='inner')
    expanded_df = expanded_df.drop(columns = [game_column]).dropna()

    return expanded_df
```

```python
fact_df = expand_fact_table(df_temp.copy(), creators_df.copy(), 'creator_id', 'creators')
fact_df_final = expand_fact_table(fact_df.copy(), developers_df.copy(), 'developer_id', 'developers')
fact_df_final_1 = expand_fact_table(fact_df_final.copy(), publishers_df.copy(), 'publisher_id', 'publishers')
```

```python
# Loi b cc sample khng tn ti trong bng DIM v bng trung gian
def final_fix_data(df_1, df_2, col_fix):
    sample_need_1 = list(set(df_1[col_fix]).difference(set(df_2[col_fix])))
    df_1 = df_1[~df_1[col_fix].isin(sample_need_1)].reset_index(drop = True)

    sample_need_2 = list(set(df_2[col_fix]).difference(set(df_1[col_fix])))
    df_2 = df_2[~df_2[col_fix].isin(sample_need_2)].reset_index(drop = True)
    return df_1, df_2
```
